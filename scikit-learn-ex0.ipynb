### 선형 회귀 모델

```python
# 선형회귀를 위한 데이터셋을 직접 만들어보자
# y=2x+1
# x = [1,2,3,4,5]
# y= [3,5,7,9,11]

import numpy as np
import matplotlib.pyplot as plt

n = 10
x_total = np.arange(1 , n+1).reshape(-1,1)
# x_total.shape
x_total
```

```python
#  y = 2x+1
y_total = 2 * x_total + 1
y_total
```

```python
randValue = np.random.randn(n)*3
randValue
```

```python
# randValue = np.random.randn(n)*3
randValue =randValue.reshape(-1,1)
y_total = 2*x_total+1+randValue
y_total
```

```python
# y = 2*x+1+난수(3사이의 난수 10개)
randValue = np.random.randn(n)*3
# reshape 안했을 때 나오는 결과는 무엇?
# reshape 하는 이유는 그냥 +n 하면 브로드캐스팅 연산. 결과2차원행렬에 2차원행렬 더할거라서 reshape
# randValue = randValue.reshape(-1,1)
y_total = 2*x_total+1+randValue
y_total
```

```python
# x_total, y_total 데이터를 산점도 그래프로 그리기
import matplotlib.pyplot as plt
plt.scatter(x_total, y_total)
plt.show()
```

### 데이터셋을 나누기

```python
# x_total, y_total을 나누기
from sklearn.model_selection import train_test_split

# 나누는 비율 0.8:0.2
x_train, x_test, y_train, y_test = train_test_split(x_total,y_total, test_size=0.2, random_state=42)
```

```python
x_train, x_test
```

```python
y_train, y_test
```

### 선형회귀 모델을 사용하여 학습, 평가

```python
# sklearn의 선형회귀 모델을 사용
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
# 1. 모델 생성(선언)
model = LinearRegression()
# model = LogisticRegression()

#2. 모델 학습
model.fit(x_train, y_train)
# model.fit(x_train.values, y_train.values.ravel())
#3. 모델 예측
y_pred = model.predict(x_test)

#4. 모델 평가
# 이건 놉
# from sklearn.metrics import accuracy_score
# accuracy = accuracy_score(y_test, y_pred)
# print("Accuracy : ", accuracy)

from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
# 오차값이 얼마나 큰지
```
