```python
# Braund, Mr. Owen Harris	
# 0    ,  ,  
train_df['Name'][0].split(',')[1].split('.')[0].strip()
```

```python
#   ( )     
# test = lambda x : x+2
# test(2)

#     title
# x = train_df['Name'][0]
conversion_rare = lambda x : x.split(',')[1].split('.')[0].strip()
# mapping  ()     .
train_df['Title'] = train_df['Name'].map(conversion_rare)
# conversion_rare(train_df['Name'][0])
titles = train_df['Title'].unique()
print(titles)
# Mrs = Mme, Miss = Ms = msll  .  ->    
```

```python
#      

# ['Mr' 'Mrs' 'Miss' 'Master' 'Don' 'Rev' 'Dr' 'Mme' 'Ms' 'Major' 'Lady'
# 'Sir' 'Mlle' 'Col' 'Capt' 'the Countess' 'Jonkheer']
#     .

train_df['Title'] = train_df['Title'].replace('Mlle', 'Miss')
train_df['Title'] = train_df['Title'].replace('Ms', 'Miss')
train_df['Title'] = train_df['Title'].replace('Mme', 'Mrs')

Rare = ['Lady','the Countess','Countess','Capt', 'Master',
        'Col','Don','Dr','Major','Rev','Sir','Jonkheer', 'Dona']

for each in Rare:
    train_df['Title'] = train_df['Title'].replace(each, 'Rare')
    
train_df['Title'].unique()

```

```python
# get_dummies  ['Pclass','Sex','Age','Embarked','Title']  
# pc, sex,emb,age,titles

train_df2 = pd.get_dummies(train_df, columns=['Pclass','Sex','Age_cat','Embarked','Title'])
```

#### 4. ()

```python
# train_df2.info()
print(train_df2.columns.tolist())
```

```python
train_df2 = train_df2.drop(columns=['Name'])
train_df2.columns.tolist()
```

```python
#   
plt.figure(figsize=(15, 15))
sns.heatmap(train_df2.corr(numeric_only=True), linewidths=0.01, square=True,
            annot=True, cmap=plt.cm.coolwarm, linecolor="white")
plt.title('Correlation between features')
plt.show()
```

```python
#              

# survived   +-0.2   
# train_df2.loc[train_df2.corr(numeric_only=True)['Survived'].abs().sort_values(ascending=False)>=0.2]
#   
colList = train_df2.corr(numeric_only=True)['Survived'].abs().sort_values(ascending=False)
train_df2.loc[:, colList]
```

```python
#   (train/tst)

# from sklearn.model_selection import train_test_split
# x_train, x_test, y_train, y_test = train_test_split(train_df2.drop('Survived'), train_df2['Survived'], test_size=0.2, random_state=11)

#  (train/test)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_df2.drop('Survived', axis=1), train_df2['Survived'], test_size=0.2, random_state=42)
```

####    Decision Tree  

```python
# De
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)

print(accuracy)
```
